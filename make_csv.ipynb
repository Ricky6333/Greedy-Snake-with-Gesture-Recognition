```python
import os,csv
import pandas as pd
```

```python
x = []
y = []
# [1,2,3,4] == [up,down,right,left]
```

```python
name_list = ["up","down","right","left"]

for name in name_list:
    # 路徑設置
    path_str = f"./Hand_data/train_data/{name}/"
    path = os.path.join(path_str)
    print(name)

    # 取出資料夾中檔案
    for file in os.listdir(path):
        x_file = os.path.join( f"/{name}/", file)
        x.append(x_file) # 加入list

        if name == "up":
            y.append(0) # up=0
        elif name == "down":
            y.append(1)
        elif name == "right":
            y.append(2)
        elif name == "left":
            y.append(3)
```

```python
dict = {'x':x, 'y':y}   # 做成字典
df = pd.DataFrame(dict)
df.to_csv("train.csv")  # 存成csv
```
